**Question 7 :** A company wants to calculate performance bonuses for employees based on their performance
ratings and sales achieved.

Task:
*   If PERFORMANCE RATING _"EXCELLENT" and SALES>=100000 , Bonus = 20% of Sales.
*   If PERFORMANCE RATING ="GOOD"and SALES>=80000 , Bonus = 15% of Sales.
*   Else, Bonus = 10% of Sales.

**Answer:-**

image.png

**Formula Used:-** =IF(AND($C48="Excellent",$B48>100000),"Bonus = 20% of Sales",IF(AND($C48="Good",$B48>80000),"Bonus = 15% of Sales","10% of Sales"))
